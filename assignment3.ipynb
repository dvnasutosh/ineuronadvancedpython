3. **Describe the significance of DASK.**
   Dask is a Python library for scalable and efficient handling of large datasets and complex computations. It enables parallel computing and seamless integration with popular Python libraries like pandas and NumPy. With Dask, you can work with datasets larger than memory, leverage lazy evaluation for optimized performance, and scale computations from a single machine to a distributed cluster. It is widely used in big data processing, machine learning, and analytics workflows.

4. **Describe the functions of DASK.**

The main functions of Dask are as follows:
1. Dask enables scalable parallel computing by breaking computations into smaller tasks and executing them across multiple cores or machines.
2. Lazy evaluation in Dask optimizes memory usage and schedules computations efficiently.
3. Dask seamlessly integrates with popular Python libraries like pandas and NumPy, providing familiar APIs for working with large datasets.
4. Dask supports out-of-core computing, allowing processing of datasets larger than available memory.
5. Dask facilitates distributed computing, executing computations across multiple machines in a cluster.
6. Dask is widely used in machine learning and big data analytics workflows, enabling distributed training and processing of large datasets.


5. **Describe Cassandra's features.**
   1. Cassandra offers horizontal scalability, allowing it to handle large amounts of data by adding more machines to the cluster.
   2. It provides high availability through data replication, ensuring data accessibility even in the event of node failures.
   3. Cassandra is fault-tolerant, with mechanisms in place to handle node failures without compromising data integrity or availability.
   4. Its distributed architecture enables efficient data distribution, improving performance, scalability, and fault tolerance.
   5. Cassandra has a flexible data model that supports structured, semi-structured, and unstructured data.
   6. Consistency levels in Cassandra are tunable, allowing developers to balance data consistency and performance.
   7. It is optimized for fast writes and reads, leveraging log-structured storage and parallel processing.
   8. Cassandra offers various replication strategies and supports clustering for customized data distribution and replication.
   9. Tunable durability allows users to configure how often data is persisted to disk for durability.
   10. It includes built-in caching mechanisms for improved read performance by caching frequently accessed data in memory.
